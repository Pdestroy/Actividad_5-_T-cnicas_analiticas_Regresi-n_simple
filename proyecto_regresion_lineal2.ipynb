# Proyecto: Regresión Lineal con Machine Learning

En este proyecto se implementara el uso de un modelo de **Regresión Lineal** para aprender la relación entre grados Celsius y Fahrenheit.

La regresión lineal es uno de los algoritmos más simples y utilizados en Machine Learning, ya que permite modelar la relación entre una variable de entrada y una variable de salida mediante una línea recta.

## ¿Qué es la Regresión Lineal?

La regresión lineal es un algoritmo de aprendizaje automático que permite encontrar la relación entre dos variables.

Se representa mediante la ecuación:

Y = mX + b

Donde:
- X = variable independiente
- Y = variable dependiente
- m = pendiente
- b = intercepto

El modelo aprende estos valores a partir de los datos de entrenamiento.

## Importar librerías necesarias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Crear el conjunto de datos

```python
# =========================
# Cargar datos desde CSV
# =========================
df = pd.read_csv("datos_temperatura.csv")

# Ver datos
print(df.head())

# =========================
# Preparar datos
# =========================
celsius = df['Celsius'].values.reshape(-1,1)
fahrenheit = df['Fahrenheit'].values
```

## Visualizar los datos

```python
plt.scatter(celsius, fahrenheit)
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.title('Relación entre Celsius y Fahrenheit')
plt.show()
```

## Crear el modelo de Regresión Lineal

```python
modelo = LinearRegression()
```

## Entrenar el modelo

```python
modelo.fit(celsius, fahrenheit)
```

## Realizar una predicción

```python
prediccion = modelo.predict([[25]])
print('25 grados Celsius equivalen aproximadamente a:', prediccion[0], 'grados Fahrenheit')
```

## Graficar la línea de regresión

```python
predicciones = modelo.predict(celsius)

plt.scatter(celsius, fahrenheit)
plt.plot(celsius, predicciones)

plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.title('Modelo de Regresión Lineal')

plt.show()
```

## Conclusión

La regresión lineal permite identificar la relación entre variables y realizar predicciones basadas en datos históricos.
